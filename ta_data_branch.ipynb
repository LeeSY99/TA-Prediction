```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("대구.csv")
df2 = pd.read_csv("전국.csv")
```

```python
# Creating df3 for rows with '광역시'
df3 = df2[df2['광역시'].notna()].drop(columns=['도', '시군'])

# Creating df4 for rows without '광역시'
df4 = df2[df2['광역시'].isna()].drop(columns=['광역시', '구'])
```

```python
df5 = pd.concat([df1, df3], ignore_index=True)

df5
```

```python
df4.to_csv('시군.csv', index=False)
df5.to_csv('광역시.csv', index=False)
```
